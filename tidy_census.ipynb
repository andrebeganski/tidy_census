```python
##Import the pandas
import pandas as pd
```

```python
##Read the CSV
df = pd.read_csv("census.csv")
df
```

```python
##dropping columns that are not needed
drop_df = df.drop(["state", "state fips"], axis=1)
drop_df
```

```python
##Melting years into the same variable
melt_df = drop_df.melt(id_vars = ["commodity", "domain category", "data item"],
                  var_name = "year",
                  value_name = "amount")
melt_df
```

```python
##Splitting the column that contains multiple variables
melt_df[["commodity_type", "data_type"]] = melt_df["data item"].str.split("-", n=1, expand = True,)
melt_df
```

```python
##dropping the column with missing data
melt_df = melt_df.drop(["domain category"], axis = 1)
melt_df
```

```python
##dropping data item column because it's redundant
melt_df = melt_df.drop(["data item"], axis = 1)
melt_df
```

```python
##reorganizing columns in a way that makes more sense
melt_df = melt_df.reindex(columns=["commodity", "commodity_type", "data_type", "year", "amount"])
melt_df
```

```python
##Sorting values by commodity
melt_df = melt_df.sort_values(by = ["year", "commodity"], ascending = True, ignore_index = True)
melt_df
```
